#### 5. Find the movies that have more than one genre - 2.5 marks

```python
# From exploring the data, we know the genre columns contain booleans that indicate whether
# a movie is of that genre, or not. So, finding the movies with more than one genre is done
# by summing the booleans, and dropping any results with a count of 1 or less.

# Drop the movie id, movie title, and release date fields, and sum the resulting dataframe by
# rows (axis=1). Name the results column 'Genre Count', and convert the result to a new temporary
# working dataframe df.
df = pd.DataFrame((df_Movies.drop(labels=['movie id','movie title','release date'], axis=1).sum(axis=1, numeric_only=True)),columns=['Genre Count'])

# Find the indexes of all rows in the dataframe where Genre Count is less than or equal to one,
# then drop these rows from the dataframe. The result is a dataframe with the
# Genre Count containing the sum of the genres observed for that movie, with all movies
# of only 1 genre (or none, in case there are observations in the dataset with no genre) dropped.
indexNames = df[df['Genre Count'] <= 1].index
df.drop(indexNames,inplace=True)

# Now join the results dataframe to the movie data to produce a result showing the movie id,
# the title of the movie, and the number of genres. Drop the same indexes from the source Movie
# dataframe (to exclude the movies with <= 1 genre) and create a new dataframe with the movie id
# and movie title columns, left joined to the dataframe containing the genre counts.
df_MultiGenreMovies = (pd.DataFrame((df_Movies.drop(indexNames))[['movie id','movie title']])).join(df,how='left')

# Reset the index for the new dataframe
df_MultiGenreMovies.reset_index(inplace=True)
df_MultiGenreMovies.drop(labels=['index'], axis=1, inplace=True)


print(df_MultiGenreMovies.shape)
print('\n')
print(df_MultiGenreMovies)
df_MultiGenreMovies.head()
```

```python
# Observations:
# There are 849 movies in the datasource that are associated with more than one genre.
```

### 8. Visualize how popularity of genres has changed over the years - 10 marks

Note that you need to use the number of releases in a year as a parameter of popularity of a genre

Hint 

1: you need to reach to a data frame where the release year is the index and the genre is the column names (one cell shows the number of release in a year in one genre) or vice versa.
Once that is achieved, you can either use univariate plots or can use the heatmap to visualise all the changes over the years 
in one go. 

Hint 2: Use groupby on the relevant column and use sum() on the same to find out the nuumber of releases in a year/genre.  

```python
# The approach here is as suggested at point 1 above, which is to produce a dataframe
# where the release year is the index, and the genres are in column names with each value
# showing the total releases of that genre for the year.

# Create a temporary dataframe from the Movies data
df = df_Movies.copy()

# Prepare the data by dropping any movies with "unknown" genre
indexNames = df[df['unknown'] == 1].index
df.drop(indexNames, inplace=True)

# Replace all the release dates with the corresponding release year
df['release date'] = df_Movies['release date'].str.extract('(?P<year>\d{4})').astype(int)

# Drop the superfluous columns
df.drop(labels=['movie id','movie title','unknown'], axis=1, inplace=True)

# Use groupby on the year values now in 'release date' to create a DataFrameGroupBy object, 
# then sum the columns to produce the total releases for the year, and convert to DataFrame
df_GenrePopularity = pd.DataFrame(df.groupby('release date').sum())

print(df_GenrePopularity.info())
print('')
df_GenrePopularity
```

```python
# Observations:
# Have successfully created the required dataframe with release years as the index, and columns
# for each genre with the values being the number of times that genre was observed for any movie
# released in each year. Before plotting, a quick check shows there are no null values in
# any of the genre columns, and the shape shows there are 71 different release years observed
# in the dataset.
```

```python
# Visualise using a heatmap to show the whole dataset.
plt.figure(figsize=(10,8))
sns.heatmap(df_GenrePopularity.transpose(), cmap="Blues");
```

```python
# Observation:
# Perhaps unsurprisingly, given the observations of the Release Year distribution plot above,
# the fact that there is a concentration of release years after 1990 has the effect of
# hiding any trends from before 1990 that might otherwise be interesting to observe. One
# option to surface more of these trends is to instead present the heatmap based on the log10
# of the dataframe.
```

```python
# Use the apply function to produce a new dataframe which is the log of the dataframe visualised
# in the heatmap above
df_GenrePopularityLog = (df_GenrePopularity.apply(np.log10))

# Then produce the heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df_GenrePopularityLog.transpose(), cmap="Blues");
```

```python
# Observations:
# Mapping the log of the dataframe still shows the concentration of releases after 1990, but
# now more of the trends for individual genres before 1990 are visible. Comedy and Drama
# stand out as having been more consistently popular compared to other genres, with the trends
# for genres like Action, Adventure, and Sci-Fi now visible from earlier.
```

### 9. Find the top 25 movies according to average ratings such that each movie has number of ratings more than 100 - 10 marks

Hint : 
1. First find the movies that have more than 100 ratings(use merge, groupby and count). Extract the movie titles in a list.
2. Find the average rating of all the movies and sort them in the descending order. You will have to use the .merge() function to reach to a data set through which you can get the names and the average rating.
3. Use isin(list obtained from 1) to filter out the movies which have more than 100 ratings.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
# My approach is to use the Ratings data and first prepare a DataFrameGroupBy
# which is grouped by movie id, then create a DataFrame using the .count()
# and .mean() values, then drop rows with fewer than 100 ratings. This DataFrame
# is then inner merged with the Movies data - this does the filtering in one step

# Create a working copy of the Ratings DataFrame
df = df_Ratings.copy()

# Drop the superfluous columns
df.drop(labels=['user id','timestamp'], axis=1, inplace=True)

# Create a DataFrameGroupBy using groupby on 'movie id', and create a new DataFrame
# populated with the .count() value - gives the number of times the 'movie id' appears
# in the Ratings observations
df_TopRatings = pd.DataFrame(df.groupby('movie id').count())

# Now add the average rating to the new DataFrame, and tidy up the labels
df_TopRatings['average rating'] = df.groupby('movie id').mean()
df_TopRatings.rename(columns = {'rating':'number of ratings'}, inplace=True)

# Find the indexes for where there are fewer than 100 ratings and drop them
indexNames = df_TopRatings[df_TopRatings['number of ratings'] < 100].index
df_TopRatings.drop(indexNames, inplace=True)

# Create a new DataFrame with an inner merge of the original Movies data and the
# new DataFrame which now contains movie id, number of ratings, and average rating
# only for movies with more than 100 ratings. Inner merge on 'movie id' excludes all
# movies with fewer than 100 ratings.
df_TopMovies = pd.merge(df_Movies, df_TopRatings, how='inner', on='movie id')

# Sort the DataFrame by 'average rating', with ascending=False so the sorting is highest
# to lowest average rating
df_TopMovies.sort_values(by='average rating', ascending=False, inplace=True)

# Create the finished DataFrame by extracting the first 25 rows
df_Top25Movies = df_TopMovies.iloc[0:25]

# This now contains the top 25 movies by sorted by average rating, for only movies
# with more than 100 ratings
df_Top25Movies
```

```python
# My approach is to produce a dataframe that has gender ('M' or 'F') as the index, and genre
# as the columns, with the values being the number of times anyone of each gender has given a
# rating for a movie which is of each genre.

# First, create a dataframe that adds gender to each movie id that appears in the Ratings data
# by doing an inner merge of Ratings and Users on user id, then creating a new dataframe
# from the resulting gender and movie id
df_GenderedRatings = (pd.merge(df_Ratings, df_Users, how='inner', on='user id'))[['gender','movie id']]
print(df_GenderedRatings.shape)

# Second, create a new dataframe from the Movies data, inner merged on movie id to the gendered
# ratings data. Then drop all the columns that are not being summed, then groupby gender and
# sum the results into the final dataframe
df_GenderedGenres = pd.merge(df_GenderedRatings, df_Movies, how='inner', on='movie id')
print(df_GenderedGenres.shape)
df_GenderedGenres.drop(labels=['movie id','movie title','release date','unknown'],axis=1,inplace=True)
df_TotalGenderedGenres = pd.DataFrame(df_GenderedGenres.groupby('gender').sum())

df_TotalGenderedGenres
```

```python
# Observations:
# This successfully creates the desired dataframe, which shows the number of times users of
# each gender have watched a movie which is of each genre.

# One data quality issue emerges from exploratory analysis in preparing the dataframes,
# which is that there are 9 rows in the Ratings dataframe that contain the 
# movie id '267', which does not appear in the Movies data. This is observed on the
# second merge where the Ratings data (now containing gender) is merged with the
# Movies data using an inner merge on movie id. The shape result shows the row count
# dropping to 99991. Although not elaborated in the answer above, below is the code that
# uses 'isin' to identify which ratings have a movie id which does not appear in the Movie data.

#df = pd.DataFrame(df_Ratings['movie id'].isin(df_Movies['movie id']))
#indexNames = df[df['movie id'] == False].index 
#s_DroppedRatings = (pd.DataFrame((df_Ratings['movie id'].iloc[indexNames]))).groupby('movie id').sum()
#s_DroppedRatings
```

```python
# The next step is to prepare four metrics which will be used in the analysis - the number of
# users of each gender in the Users data, and the total number of ratings (movies watched)
# for each gender.
df = df_Users.copy()
df.drop(labels=['age','occupation','zip code'],axis=1,inplace=True)
countFemale = ((pd.DataFrame(df.groupby('gender').count())).loc['F'])['user id']
countMale = ((pd.DataFrame(df.groupby('gender').count())).loc['M'])['user id']

reviewsByFemale = df_TotalGenderedGenres.loc['F'].sum()
reviewsByMale = df_TotalGenderedGenres.loc['M'].sum()

print ("Number of women: " + str(countFemale))
print ("Number of men: " + str(countMale))
print ("Ratings by women: " + str(reviewsByFemale))
print ("Ratings by men: " + str(reviewsByMale))
```

```python
# Observations:
# As had been noted in the countplot analysis above, there are more than twice the number of 
# men than women in this user cohort.
```

```python
# Each of the three hypothesis above are considered in turn. My approach is to consider two
# metrics - one is the "Average Number Watched" by individuals of each gender of each genre, 
# which normalises the comparison as there are many more men than women in the cohort. That is,
# how many movies of each genre does an individual of each gender watch on average. The second
# metric is the "Rate Watched", which compares each genre to other genres for that gender, giving
# a basic sense of 'preference' for one genre over another.
```

```python
# Hypothesis: Men watch more drama than women

genre = "Drama"

df_Comparison = pd.DataFrame(index=['M','F'], columns=['Average Number Watched', 'Rate Watched'])
df_Comparison.loc['M', 'Average Number Watched'] = ((df_TotalGenderedGenres.loc['M'])[genre])/countMale
df_Comparison.loc['M', 'Rate Watched'] = ((df_TotalGenderedGenres.loc['M'])[genre])/reviewsByMale
df_Comparison.loc['F', 'Average Number Watched'] = ((df_TotalGenderedGenres.loc['F'])[genre])/countFemale
df_Comparison.loc['F', 'Rate Watched'] = ((df_TotalGenderedGenres.loc['F'])[genre])/reviewsByFemale

df_Comparison
```

```python
# Observations:
# This result shows that while the average male reviewers in this cohort has watched
# 43 drama movies, which is more that the 40 watched by the average female reviewer, male
# reviewers watch Drama at a lower rate (18%) compared to other genres than females (20%).

# The hypothesis 'Men watch more drama than women' is not supported by these results. While the
# average male watches more Drama, men also watch more movies overall so this effect can't be
# excluded. The metric that is less sensitive to the overall rate of watching gives the opposite
# result, so the hypothesis can't be supported.
```

```python
# Hypothesis: Women watch more Sci-Fi than men

genre = "Sci-Fi"

df_Comparison = pd.DataFrame(index=['M','F'], columns=['Average Number Watched', 'Rate Watched'])
df_Comparison.loc['M', 'Average Number Watched'] = ((df_TotalGenderedGenres.loc['M'])[genre])/countMale
df_Comparison.loc['M', 'Rate Watched'] = ((df_TotalGenderedGenres.loc['M'])[genre])/reviewsByMale
df_Comparison.loc['F', 'Average Number Watched'] = ((df_TotalGenderedGenres.loc['F'])[genre])/countFemale
df_Comparison.loc['F', 'Rate Watched'] = ((df_TotalGenderedGenres.loc['F'])[genre])/reviewsByFemale

df_Comparison
```
